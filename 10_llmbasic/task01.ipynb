```python
# キャラクターの設定（システムプロンプト）を定義
SYSTEM_SETTING = "あなたは、語尾に『～だにゃ』をつけて話す、少し生意気な猫型ロボットのキャラクターとして振る舞ってください。"

# メッセージを格納するリスト（最初にシステムプロンプトをセットしておく）
messages = [{"role": "system", "content": SYSTEM_SETTING}]

while(True):
    # ユーザーからの質問を受付
    message = input("メッセージを入力:")
    # 質問が入力されなければ終了
    if message.strip() == "":
        break
    print(f"質問: {message}")

    # メッセージにユーザーからの質問を追加
    messages.append({"role": "user", "content": message.strip()})

    # やりとりが増えた際の履歴管理
    # システムプロンプト(1件) + 会話履歴(最大8件) = 9件を超えたら、一番古い「会話」を削除
    if len(messages) > 9:
        # index 0 はシステムプロンプトなので、index 1（最古のユーザー発言）を削除する
        messages.pop(1)

    # APIへリクエスト
    stream = client.chat.completions.create(
        model=MODEL_NAME,
        messages=messages,
        stream=True,
    )

    # 言語モデルからの回答を表示
    response_message = ""
    for chunk in stream:
        if chunk.choices:
            next_content = chunk.choices[0].delta.content
            if next_content is not None:
                response_message += next_content
                print(next_content, end='', flush=True)

    # メッセージに言語モデルからの回答を追加
    messages.append({"role": "assistant", "content": response_message})
    print() # 改行用

print("\n---ご利用ありがとうございました！---")
```

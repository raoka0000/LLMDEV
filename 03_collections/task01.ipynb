```python
grades = [85, 90, 78, 92, 88]

average = sum(grades) / len(grades)

highest = max(grades)
lowest = min(grades)

above_avg_count = 0
for score in grades:
    if score >= average:
        above_avg_count += 1

print(f"平均点: {average}")
print(f"最高点: {highest}")
print(f"最低点: {lowest}")
print(f"平均以上の学生数: {above_avg_count}")
```

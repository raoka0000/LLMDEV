```python
# TODOリストを格納するリスト
filename = "todo_list.txt"
try:
    # ファイルが存在したら内容を読込
    with open(filename, "r", encoding="utf-8") as file:
        todo_list = [line.strip() for line in file]
except FileNotFoundError:
    # ファイルが存在しない場合は新しいリストを作成
    todo_list = []

# メニュー表示
print("【TODOリスト】")

while True:
    # メニューに「3:削除」を追加
    print("番号を入力してください[1:追加、2:表示、3:削除、4:保存と終了]:", end="", flush=True)
    choice = input("")

    if choice == "1":
        print("1:追加", flush=True)
        add_todo = input("追加するTODO: ")
        todo_list.append(add_todo)
        
    elif choice == "2":
        print("2:表示", flush=True)
        for i, todo in enumerate(todo_list):
            print(f"{i}. {todo}")
            
    elif choice == "3":
        print("3:削除", flush=True)
        # リストが空の場合は削除処理を行わない
        if not todo_list:
            print("削除するTODOがありません。")
            continue
            
        # 現在のリストを表示
        for i, todo in enumerate(todo_list):
            print(f"{i}. {todo}")
        
        try:
            # 削除したい番号の入力を受け取る
            delete_num = input("削除する番号: ")
            idx = int(delete_num)
            
            # 指定された番号がリストの範囲内か確認
            if 0 <= idx < len(todo_list):
                removed = todo_list.pop(idx)
                # 削除成功時のメッセージ（任意）
                # print(f"「{removed}」を削除しました。")
            else:
                print("無効な番号です。")
                
        except ValueError:
            # 数字以外が入力された場合
            print("無効な番号です。")

    elif choice == "4":
        print("4:保存と終了", flush=True)
        with open(filename, "w", encoding="utf-8") as file:
            file.writelines(f"{todo}\n" for todo in todo_list)
        print("TODOリストを保存して終了します。", flush=True)
        break
    else:
        print("1-4の数字を入力してください。", flush=True)
```

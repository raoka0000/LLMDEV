```python
words = []

# 1. 英単語の入力を繰り返す
while True:
    word = input("英単語を入力してください：")
    # 未入力（エンターのみ）の場合はループを抜ける
    if word == "":
        break
    words.append(word)

# 2. 単語をABC順に並べ替えて表示
words.sort()
print(f"入力した英単語：{words}")

# 3. アルファベットごとの数をカウントする
char_counts = {}

for word in words:
    for char in word:
        # 辞書を使って文字をカウント（既にあれば+1、なければ1をセット）
        if char in char_counts:
            char_counts[char] += 1
        else:
            char_counts[char] = 1

# 4. アルファベット順に結果を表示
# 辞書のキー（文字）を取り出してソートする
for char in sorted(char_counts.keys()):
    print(f"{char}が{char_counts[char]}個ありました")
```

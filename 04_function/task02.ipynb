```python
import urllib.request
import datetime

def save_url_content(url):
    # 1. 指定したURLからデータを取得
    with urllib.request.urlopen(url) as response:
        # UTF-8でデコードし、先頭500文字を取得
        content = response.read().decode('utf-8')
        short_content = content[:500]
    
    # 2. 現在の日時を取得してファイル名を作成
    now = datetime.datetime.now()
    # strftimeで「年月日_時分秒」のフォーマットに整える
    filename = now.strftime("%Y%m%d_%H%M%S") + ".txt"
    
    # 3. ファイルへ保存
    with open(filename, "w", encoding="utf-8") as file:
        file.write(short_content)
    
    # 4. 作成したファイル名を返す
    return filename

# --- 呼び出し側の処理 ---
url = input("URLを入力してください: ")
file_name = save_url_content(url)
print(f"Content saved in: {file_name}")
```

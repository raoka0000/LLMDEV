```python
# 1. 元金と年利をユーザーから入力（小数にも対応できるようfloat型に変換）
principal = float(input("元金を入力してください: "))
rate = float(input("年利を入力してください: "))

# 2. 計算用の変数に初期値をセット
amount = principal

# 3. for文を使って10年分繰り返し計算を行う
for year in range(1, 11):
    # 前年の金額に (1 + 利率/100) を掛けて更新
    amount = amount * (1 + (rate / 100))

# 4. 結果を表示（:.2f で小数点以下2桁までにフォーマット）
print(f"10年後の金額: {amount:.2f}")
```

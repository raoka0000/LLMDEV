```python
price_ex = int(input("税抜価格を入力してください: "))

tax_rate = 1.1
price_inc = int(price_ex * tax_rate)

if price_inc >= 2000:
    shipping_fee = 0
    shipping_msg = "無料"
else:
    shipping_fee = 350
    shipping_msg = "350円"

total_price = price_inc + shipping_fee

print(f"送料は{shipping_msg}です")
print(f"送料込税込価格は{total_price}円です")
```
